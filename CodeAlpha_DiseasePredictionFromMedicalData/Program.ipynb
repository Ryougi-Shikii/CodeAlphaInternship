```python

from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np

path = "https://drive.google.com/uc?export=download&id=1Tgk3EwGUeX-69NA7PeSjXqVbaAjUOGAl"
df = pd.read_csv(path)

x_train, x_test, y_train, y_test = train_test_split(df.drop(['Outcome', 'DiabetesPedigreeFunction', 'Pregnancies',"SkinThickness"], axis=1), df['Outcome'], test_size=0.3, random_state=42)

mlp=RandomForestClassifier(n_estimators=10 ,random_state=42)
mlp.fit(x_train, y_train)

y_pred = mlp.predict(x_test)
accuracy =accuracy_score(y_test,  y_pred)

import pickle
#saving the parameters!
with open('Diabetes.pkl', 'wb') as f:
    pickle.dump(mlp, f)

#(Input:['Glucose', 'BloodPressure', 'Insulin', 'BMI', 'Age']
#LogisticRegression -> Accuracy: 0.7316017316017316
#MLPClassifier -> Accuracy: 0.7229437229437229
#RandomForestClassifier -> Accuracy: 0.7445887445887446
```
